```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Download sample dataset
!wget -q https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv -O sales_data.csv
```

```python
# Load dataset
df = pd.read_csv("sales_data.csv")

# Display first 5 rows
df.head()
```

```python
# Shape of dataset (rows, columns)
print("Shape:", df.shape)

# Column names
print("Columns:", df.columns.tolist())

# Info about dataset
print("\nDataset Info:")
print(df.info())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Clean up column names
df.columns = df.columns.str.strip().str.replace('"', '')

# Group by weight and get average Height
grouped_data = df.groupby("Weight(Pounds)")["Height(Inches)"].mean()
print(grouped_data)
```

```python
# Bar chart for average height by weight
grouped_data.plot(kind="bar", figsize=(12, 6), color="skyblue")

plt.title("Average Height by Weight", fontsize=16)
plt.xlabel("Weight (Pounds)", fontsize=14)
plt.ylabel("Height (Inches)", fontsize=14)
plt.xticks(rotation=90)  # Rotate labels so they don't overlap
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
